```python
import numpy as np # only need this once per program
import matplotlib as mpl
from matplotlib import pyplot as plt 
```

```python
# 1. Prepare data
x = np.arange(0.5, 5.5, 1.0)
y = np.exp(-x)
yerr_values = 0.2*y  # Constant error in the y-direction
xerr_values = 0  # Constant error in the x-direction

# 2. Create Figure and Axes
fig, ax = plt.subplots(figsize=(7, 4)) # fig is the Figure, ax is the Axes object

# 3. Plot error bars using the Axes method
ax.errorbar(x, y, xerr= xerr_values, yerr=yerr_values, fmt='.', color='black', capsize=3, elinewidth=2)

# Optional: Add titles and labels using Axes methods
ax.set_title('Object-Oriented Error Bar Plot')
ax.set_xlabel('X values')
ax.set_ylabel('Y values with error')
ax.set_yscale('log')

# 4. Display the plot
plt.show()
```
